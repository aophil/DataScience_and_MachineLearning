## Getting to know our data

- userId: Unique number given to each user.

- tier: Tier of the city in which the user is residing.

- gender: Categorical feature representing the gender of the user. 1 represents male and 2 represents female.

- following_rate: Number of accounts followed by the user(feature is normalized)

- followers_avg_age: Average of age groups of all the followers of the user.

- following_avg_age: Average of age groups of all the accounts followed by the user.

- max_repetitive_punc: Maximum repititive punctuations found in the bio and comments of the user.

- num_of_hashtags_per_action: Average nubmer of hashtags used by the user per comment.

- emoji_count_per_action: Average number of emojis used by the user per comment

- punctuations_per_action: Average number of punctuations used by the user per comment.

- number_of_words_per_action: Average number of words used by the user per comment.

- avgCompletion: Average watch time completion rate of the videos.

- avgTimeSpent: Average time spent by the user on a video in seconds.

- avgDuration: Average duration of the videos that the user has watched till date.

- avgComments: Average number of comments per video watched.

- creations: Total number of videos uploaded by the user.

- content_views: Total number of videos watched.

- num_of_comments: Total number of comments made by the user (normalized)

- weekends_trails_watched_per_day: Number of videos watched on weekends per day.

- weekdays_trails_watched_per_day: Number of videos watched on weekdays per day.

- slot1_trails_watched_per_day: The day is divided into 4 slots. This feature represents the average number of videos watched in this particular time slot.

- slot2_trails_watched_per_day: The day is divided into 4 slots. This feature represents the average number of videos watched in this particular time slot.

- slot3_trails_watched_per_day: The day is divided into 4 slots. This feature represents the average number of videos watched in this particular time slot.

- slot4_trails_watched_per_day: data

- avgt2: Average number of followers of all the accounts followed by the user.

- age : The age of the user


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
train_url = pd.read_csv("https://raw.githubusercontent.com/michalis0/DataScience_and_MachineLearning/refs/heads/master/Assignements/Part%202/data/social_media_SIMF_archive.csv")
suspects_url = pd.read_csv("https://raw.githubusercontent.com/michalis0/DataScience_and_MachineLearning/refs/heads/master/Assignements/Part%202/data/suspects_social_media.csv")
```

```python
np.random.seed(42)
```

### Training a model
Select the features which by convention are called X. We will also choose the target variable which we typically call y.

You can use the features `'tier', 'gender', 'following_rate', 'followers_avg_age', 'following_avg_age', 'max_repetitive_punc', 'num_of_hashtags_per_action', 'emoji_count_per_action', 'punctuations_per_action', 'number_of_words_per_action', 'avgCompletion', 'avgTimeSpent', 'avgDuration', 'avgComments', 'creations', 'content_views', 'num_of_comments', 'weekends_trails_watched_per_day', 'weekdays_trails_watched_per_day', 'slot1_trails_watched_per_day', 'slot2_trails_watched_per_day', 'slot3_trails_watched_per_day', 'slot4_trails_watched_per_day', 'avgt2'`


```python
train = train_url.drop_duplicates().dropna()  # drop duplicates and NaN values
display(train.head())                         # Display a sample of the data
print(train.dtypes)                           # Print the data types
features = train.loc[:,['age','tier', 'gender', 'following_rate', 'followers_avg_age', 'following_avg_age', 'max_repetitive_punc', 'num_of_hashtags_per_action', 'emoji_count_per_action', 'punctuations_per_action', 'number_of_words_per_action', 'avgCompletion', 'avgTimeSpent', 'avgDuration', 'avgComments', 'creations', 'content_views', 'num_of_comments', 'weekends_trails_watched_per_day', 'weekdays_trails_watched_per_day', 'slot1_trails_watched_per_day', 'slot2_trails_watched_per_day', 'slot3_trails_watched_per_day', 'slot4_trails_watched_per_day', 'avgt2']]
display(features.describe())                  # Summary statistics
display(features.corr(numeric_only = True))   # Correlation matrix
```

Now do the train-test split, using 80% of the data as training and 20% as test. Set the random state to 42.

```python
independent = features[['age']]
dependent = features.drop(columns=['age'])
dependent_train, dependent_test, independent_train, independent_test = train_test_split(dependent, independent, test_size=0.2, random_state=42, shuffle=True)
```

Next, rescale the features in training and test data using `MinMaxScaler`.

```python
scaler = MinMaxScaler()
dependent_train = scaler.fit_transform(dependent_train)
dependent_test = scaler.transform(dependent_test)
```

## Linear regression
Now, create a linear regression model. Train it on the training data and use the model to make predictions on the test data. How is the performance of your model on the test data?

*Note: For the liner regression model, use the default attribute value as seen in the Lab.*

```python
model = LinearRegression()

model.fit(dependent_train, independent_train)

print("R\u00b2 Score of the model: ",
      round(model.score(dependent_train, independent_train), 3))

print("Intercept: ", model.intercept_[0])
model_coeff = pd.DataFrame(model.coef_.flatten(),
                     index=['tier', 'gender', 'following_rate', 'followers_avg_age', 'following_avg_age', 'max_repetitive_punc', 'num_of_hashtags_per_action', 'emoji_count_per_action', 'punctuations_per_action', 'number_of_words_per_action', 'avgCompletion', 'avgTimeSpent', 'avgDuration', 'avgComments', 'creations', 'content_views', 'num_of_comments', 'weekends_trails_watched_per_day', 'weekdays_trails_watched_per_day', 'slot1_trails_watched_per_day', 'slot2_trails_watched_per_day', 'slot3_trails_watched_per_day', 'slot4_trails_watched_per_day', 'avgt2'],
                     columns=['Coefficients multivariate model'])
model_coeff
```

and we do the predictions on the test set:

```python
predictions = model.predict(dependent_test)
```

What are the model's $MAE$, $MSE$ and $R^2$ ?

```python
mae = mean_absolute_error(independent_test, predictions)
mse = mean_squared_error(independent_test, predictions)
r2 = r2_score(independent_test, predictions)

print(f"MAE test set: {mae:0.2f};")
print(f"MSE test set: {mse:0.2f};")
print(f"R\u00b2 test set: {r2:0.2f};" )
```

## Retraining on the full dataset
Training on a split of the training dataset, as you just did, is mainly for evaluating the performance of the resulting model on the unseen data and also optimizing various hyper-parameters of the model.

We can now retrain the model on the full training data prior to making predictions.

Before predicting the age of the suspects, retrain the model on the full training dataset.

You can reuse the same features as before, but make sure those are also available on the suspects dataset.

```python
train_df = pd.read_csv("https://raw.githubusercontent.com/michalis0/DataScience_and_MachineLearning/refs/heads/master/Assignements/Part%202/data/social_media_SIMF_archive.csv")
train_clean = train_df.drop_duplicates().dropna()

feature_cols = [
    'tier', 'gender', 'following_rate', 'followers_avg_age',
    'following_avg_age', 'max_repetitive_punc', 'num_of_hashtags_per_action',
    'emoji_count_per_action', 'punctuations_per_action', 'number_of_words_per_action',
    'avgCompletion', 'avgTimeSpent', 'avgDuration', 'avgComments', 'creations',
    'content_views', 'num_of_comments', 'weekends_trails_watched_per_day',
    'weekdays_trails_watched_per_day', 'slot1_trails_watched_per_day',
    'slot2_trails_watched_per_day', 'slot3_trails_watched_per_day',
    'slot4_trails_watched_per_day', 'avgt2']

X_full_train = train_clean[feature_cols]
y_full_train = train_clean['age']
```

```python
full_scaler = MinMaxScaler()
X_full_train_scaled = full_scaler.fit_transform(X_full_train)
```

```python
final_model = LinearRegression()
final_model.fit(X_full_train_scaled, y_full_train)
```

```python
suspects_df = pd.read_csv("https://raw.githubusercontent.com/michalis0/DataScience_and_MachineLearning/refs/heads/master/Assignements/Part%202/data/suspects_social_media.csv")

X_suspects = suspects_df[feature_cols]
X_suspects_scaled = full_scaler.transform(X_suspects)

predicted_ages = final_model.predict(X_suspects_scaled)
suspects_df['predicted_age'] = predicted_ages

print("Suspects' ages predicted successfully!")
display(suspects_df[['userId', 'predicted_age']].head())
```

**Keep only the suspects between 25 and 35 years old. Use the `age` values you predicted in the previous step.**


```python
filtered_suspects = suspects_df[(suspects_df['predicted_age'] >= 25) & (suspects_df['predicted_age'] <= 35)]
suspect_ids_to_check = [410319, 785994, 123699, 757865, 255830]
remaining_suspects_df = filtered_suspects[filtered_suspects['userId'].isin(suspect_ids_to_check)]

print(remaining_suspects_df['userId'].tolist())
```
